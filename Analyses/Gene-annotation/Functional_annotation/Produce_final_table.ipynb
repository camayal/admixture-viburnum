# Produce final tables of functional annotation results

## Produce table

```python
# load second rounds for initial annotation and trichome SNPs
tables = ["./20kwindow_FunctionalAnnotation/20kwindow_blast2go_table_round2.txt",
          "./20kwindows_trichomes_FunctionalAnnotation/20kwindows_trichomes_blast2go_table_round2.txt"
]
```

```python
import pandas as pd

#load tables inline 
individual_dfs = [pd.read_csv(table, sep="\t") for table in tables]

# and concatenate them
df = pd.concat(individual_dfs, ignore_index=True)

df
```

```python
#detect duplicates 
df[df.duplicated("SeqName")]
```

```python
df.to_csv("functional_annotation_result.csv", index=False)
```
